# 練習

正負數判斷

```python
a=input("請輸入數字: ")
b=int(a)
if(b>0):
  print("正數",b)
elif(b==0):
  print(b)
else:
  print("負數",b,abs(b))
```

先選攝氏、華氏溫度再轉換 使用者輸入其他則顯示錯誤

```python
temp=input("請輸入溫標(F/C):")
if(temp=="F"):
  f=input("請輸入華氏溫度：")
  c=(int(f)-32)*(5/9)
  print(c)
elif(temp=="C"):
  c=input("請輸入攝氏溫度：")
  f=int(c)*(9/5)+32
  print(f)
else:
  print("Wrong!")
```

判斷成績:巢狀(nested) if else

```python
score=int(input("請輸入成績："))
if(score>=60):
  print("及格！")
  if(score>=90):
    print("超讚！")
  else:
    print("很好！")
else:
  print("不及格！")
  if(score>=40):
    print("還有機會！")
  else:print("gg!")
```
